```python
# cv.split, cv.merge, cv.resize, cv.add, cv.addWeighted, ROI - Region of Interest

import numpy as np
import cv2
import matplotlib.pyplot as plt


img = cv2.imread('Tiger 3.jpg', 1)
img2 = cv2.imread('Tiger 4.jpg', 1)

print(img.shape) #returns a tuple number of rows, columns and channels
print(img.size) #returns total number of pixels is accessed
print(img.dtype) #returns image datatype is obtained
b,g,r = cv2.split(img)
img = cv2.merge((b,g,r))


img = cv2.resize(img, (512,512))
img2 = cv2.resize(img2, (512,512))

dst = cv2.add(img, img2);

# cv2.imshow('image', dst)

plt.imshow(dst)

plt.show


cv2.waitKey(0)
cv2.destroyAllWindows()
```
