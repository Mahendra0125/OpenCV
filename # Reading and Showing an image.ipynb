```python
# Reading and Showing an image

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('Tiger 1.jpg', 1)
print(img)
# cv2.imshow('Image', img)

plt.imshow(img)
plt.show

m = cv2.waitKey(0)
if m == 27:
    cv2.destroyAllWindows()
elif m == ord('s'):
    cv2.imwrite('Tiger_copy.png', img)
    cv2.destroyAllWindows()
```
