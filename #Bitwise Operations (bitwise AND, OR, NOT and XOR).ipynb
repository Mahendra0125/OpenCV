```python
#Bitwise Operations (bitwise AND, OR, NOT and XOR)
#Masks - are binary images that indicates the pixel in which an operation is to be performed. 
# AND Operation

import numpy as np
import cv2
import matplotlib.pyplot as plt

img1 = np.zeros((250, 500, 3), np.uint8)
img1 = cv2.rectangle(img1, (100,20), (300,100), (255,255,444), -1)
img2 = cv2.imread('Tiger 4.jpg', 1)

img1 = cv2.resize(img1, (512,512))
img2 = cv2.resize(img2, (512,512))

bitand = cv2.bitwise_and(img2,img1)

# cv2.imshow('img1', img1)
# cv2.imshow('img2', img2)
# cv2.imshow('bitand', bitand)

plt.imshow(bitand)
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
